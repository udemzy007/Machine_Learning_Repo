```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('Constants with linear regression.csv')
```

```python
raw_data
```

```python
data = raw_data.copy()
```

```python
data['Attendance'] = data['Attendance'].map({'Yes': 1, 'No': 0})
```

```python
data
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT', 'Attendance']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
new_data = pd.DataFrame({'const': 1, 'SAT': [1700, 1670], 'Attendance': [0, 1]})
new_data = new_data[['const', 'SAT', 'Attendance']]
new_data
```

```python
new_data.rename(index = {0:'Bob', 1: 'David'})
```

```python
predictions = results.predict(new_data)
predictions
```

```python
predictiondf = pd.DataFrame({'Predictions':predictions})
joined = new_data.join(predictiondf)
joined.rename(index = {0 : 'Bob', 1: 'David'})
```
